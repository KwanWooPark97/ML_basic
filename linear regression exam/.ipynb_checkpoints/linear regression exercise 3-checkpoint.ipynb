## 다음 셀에 주어진 데이터를 이용해서 linear regression (선형 회귀) 문제를 
1. least square, 
2. batch gradient, 
3. stochastic gradient, 
4.mini-batch gradient 

## 방식으로 푸는 python 코드를 작성하시오.

```python
import numpy as np
from numpy import genfromtxt
import matplotlib.pyplot as plt
data = genfromtxt('PV.csv', delimiter=',',encoding="utf-8")

pv,temp,cloud,next_pv=np.split(data,4,axis=1)
pv=pv.flatten()
temp=temp.flatten()
cloud=cloud.flatten()
next_pv=next_pv.flatten()

plt.scatter(np.arange(len(pv)), next_pv, s = 5)
```

## 1. Least Square

$AW=y$

$W=(A^T A)^{-1}A^Ty$

$A=(A^T A)^{-1}$ 

$y=A^Ty$


```python
degree=10
W = [np.random.random() for _ in range(degree + 1)] #+1 is for W_0

def update_weights(degree, x, y):
    A = np.linalg.inv(np.array( 
        [ [sum(np.power(x, i)) for i in range(j, degree + 1 + j)] for j in range(degree + 1) ]
    ))

    return np.dot(A, np.array( [ sum(y * np.power(x, i)) for i in range(degree+1) ] ))

def predict(W, x):
    return sum([W[i] * x**i for i in range(len(W))])
    
W = update_weights(degree, pv, next_pv)
g = predict(W, pv)
```

```python
def plot_predictions(x_tr, y_tr, g):

    plt.scatter(np.arange(len(x_tr)), y_tr, s = 3, label = 'True Value')
    plt.scatter(np.arange(len(x_tr)), g,s=3,label = 'Prediction',color='r')
    plt.legend()
    
    plt.show()
    
g = sum([W[i] * pv**i for i in range(len(W))])
plot_predictions(pv, next_pv, g)
```

## 2. Batch gradient descent

$J(W)= \frac{1}{2m} \sum^{m}_{i=1}(\hat{y}^i-y^i)^2$ (Mean Squared Error)

${\partial J(W) \over \partial W_j} = \frac{1}{m} \sum^{m}_{i=1}(\hat{y}^i-y^i)x^i_j$

```python
errors = []
w1_grad = []

num_epoch=5000
learning_rate=0.0003

w1 = np.random.uniform(low=-1.0, high=1.0)
w2 = np.random.uniform(low=-1.0, high=1.0)
w3 = np.random.uniform(low=-1.0, high=1.0)
#w4 = np.random.uniform(low=-1.0, high=1.0)
#w5 = np.random.uniform(low=-1.0, high=1.0)
#w6 = np.random.uniform(low=-1.0, high=1.0)

b = np.random.uniform(low=-1.0, high=1.0)

for epoch in range(num_epoch):
    
    y_hat = w1 * pv +w2*temp +w3*cloud + b
    #y_hat = w1 * pv +w2*temp +w3*cloud +w4 * (pv**4) +w5*(temp**3) +w6*(cloud**2)+ b
    
    error = ((y_hat - next_pv)**2).mean()
    if error < 0.00001:
        break
        
    w1 = w1 - learning_rate * ((y_hat - next_pv) * pv).mean()
    w2 = w2 - learning_rate * ((y_hat - next_pv) * temp).mean()
    w3 = w3 - learning_rate * ((y_hat - next_pv) * cloud).mean()
    #w4 = w4 - learning_rate * ((y_hat - next_pv) * (pv**4)).mean()
    #w5 = w5 - learning_rate * ((y_hat - next_pv) * (temp**3)).mean()
    #w6 = w6 - learning_rate * ((y_hat - next_pv) * (cloud**2)).mean()
    
    b = b - learning_rate * (y_hat - next_pv).mean()
    
    errors.append(error)
    
    if epoch % 5 == 0:
        print("{0:2} w1 = {1:.5f}, w2 = {2:.5f}, w3 = {3:.5f},b = {4:.5f} error = {5:.5f}".format(epoch, w1, w2,w3, b, error))

print("----" * 15)
print("{0:2} w1 = {1:.1f}, w2 = {2:.1f},w3 = {3:.1f}, b = {4:.1f} error = {5:.5f}".format(epoch, w1, w2,w3,b, error))
```

```python
g=w1 * pv +w2*temp +w3*cloud + b
#g=w1 * pv +w2*temp +w3*cloud +w4 * (pv**4) +w5*(temp**3) +w6*(cloud**2)+ b
plot_predictions(pv, next_pv, g)
```

## 3. Stochastic gradient descent

$J(W)= (\hat{y}^i-y^i)^2$ (Mean Squared Error)

${\partial J(W) \over \partial W_j} = (\hat{y}-y)x_j$

```python
errors = []
w1_grad = []

num_epoch=5000
learning_rate=0.0003

w1 = np.random.uniform(low=-1.0, high=1.0)
w2 = np.random.uniform(low=-1.0, high=1.0)
w3 = np.random.uniform(low=-1.0, high=1.0)
#w4 = np.random.uniform(low=-1.0, high=1.0)
#w5 = np.random.uniform(low=-1.0, high=1.0)
#w6 = np.random.uniform(low=-1.0, high=1.0)

b = np.random.uniform(low=-1.0, high=1.0)

for epoch in range(num_epoch):
    rnd=np.random.randint(780)
    y_hat = w1 * pv[rnd] +w2*temp[rnd] +w3*cloud[rnd] + b
    #y_hat = w1 * pv[rnd] +w2*temp[rnd] +w3*cloud[rnd] +w4 * (pv[rnd]**4) +w5*(temp[rnd]**3) +w6*(cloud[rnd]**2)+ b
    
    error = ((y_hat - next_pv[rnd])**2).mean()
    #if error < 0.00001:
    #    break
        
    w1 = w1 - learning_rate * ((y_hat - next_pv[rnd]) * pv[rnd]).mean()
    w2 = w2 - learning_rate * ((y_hat - next_pv[rnd]) * temp[rnd]).mean()
    w3 = w3 - learning_rate * ((y_hat - next_pv[rnd]) * cloud[rnd]).mean()
    #w4 = w4 - learning_rate * ((y_hat - next_pv) * (pv[rnd]**4)).mean()
    #w5 = w5 - learning_rate * ((y_hat - next_pv) * (temp[rnd]**3)).mean()
    #w6 = w6 - learning_rate * ((y_hat - next_pv) * (cloud[rnd]**2)).mean()
    
    b = b - learning_rate * (y_hat - next_pv[rnd]).mean()
    
    errors.append(error)
    
    if epoch % 5 == 0:
        print("{0:2} w1 = {1:.5f}, w2 = {2:.5f}, w3 = {3:.5f},b = {4:.5f} error = {5:.5f}".format(epoch, w1, w2,w3, b, error))

print("----" * 15)
print("{0:2} w1 = {1:.1f}, w2 = {2:.1f},w3 = {3:.1f}, b = {4:.1f} error = {5:.5f}".format(epoch, w1, w2,w3,b, error))
```

```python
g=w1 * pv +w2*temp +w3*cloud + b
#g=w1 * pv +w2*temp +w3*cloud +w4 * (pv**4) +w5*(temp**3) +w6*(cloud**2)+ b
plot_predictions(pv, next_pv, g)
```

## 4. Mini-batch gradient descent

$J(W)= \frac{1}{2m} \sum^{m}_{i=1}(\hat{y}^i-y^i)^2$ (Mean Squared Error)

${\partial J(W) \over \partial W_j} = \frac{1}{m} \sum^{m}_{i=1}(\hat{y}^i-y^i)x^i_j$

```python
errors = []
w1_grad = []
mini_batch=50
num_epoch=5000
learning_rate=0.0003

w1 = np.random.uniform(low=-1.0, high=1.0)
w2 = np.random.uniform(low=-1.0, high=1.0)
w3 = np.random.uniform(low=-1.0, high=1.0)
#w4 = np.random.uniform(low=-1.0, high=1.0)
#w5 = np.random.uniform(low=-1.0, high=1.0)
#w6 = np.random.uniform(low=-1.0, high=1.0)

b = np.random.uniform(low=-1.0, high=1.0)

for epoch in range(num_epoch):
    rnd=np.random.randint(780,size=mini_batch)
    y_hat = w1 * pv[rnd] +w2*temp[rnd] +w3*cloud[rnd] + b
    #y_hat = w1 * pv[rnd] +w2*temp[rnd] +w3*cloud[rnd] +w4 * (pv[rnd]**4) +w5*(temp[rnd]**3) +w6*(cloud[rnd]**2)+ b
    
    error = ((y_hat - next_pv[rnd])**2).mean()
    #if error < 0.00001:
    #    break
        
    w1 = w1 - learning_rate * ((y_hat - next_pv[rnd]) * pv[rnd]).mean()
    w2 = w2 - learning_rate * ((y_hat - next_pv[rnd]) * temp[rnd]).mean()
    w3 = w3 - learning_rate * ((y_hat - next_pv[rnd]) * cloud[rnd]).mean()
    #w4 = w4 - learning_rate * ((y_hat - next_pv) * (pv[rnd]**4)).mean()
    #w5 = w5 - learning_rate * ((y_hat - next_pv) * (temp[rnd]**3)).mean()
    #w6 = w6 - learning_rate * ((y_hat - next_pv) * (cloud[rnd]**2)).mean()
    
    b = b - learning_rate * (y_hat - next_pv[rnd]).mean()
    
    errors.append(error)
    
    if epoch % 5 == 0:
        print("{0:2} w1 = {1:.5f}, w2 = {2:.5f}, w3 = {3:.5f},b = {4:.5f} error = {5:.5f}".format(epoch, w1, w2,w3, b, error))

print("----" * 15)
print("{0:2} w1 = {1:.1f}, w2 = {2:.1f},w3 = {3:.1f}, b = {4:.1f} error = {5:.5f}".format(epoch, w1, w2,w3,b, error))
```

```python
g=w1 * pv +w2*temp +w3*cloud + b
#g=w1 * pv +w2*temp +w3*cloud +w4 * (pv**4) +w5*(temp**3) +w6*(cloud**2)+ b
plot_predictions(pv, next_pv, g)
```
