```python
import numpy as np
import matplotlib.pyplot as plt
import math
from mpl_toolkits.mplot3d import Axes3D
np.random.seed(42)

#m=500
num_data = 500;

X = 10*np.random.rand(num_data,2)-5
y = np.sin(X[:,0]) + 0.5*np.sin(X[:,0]*X[:,1]) + 0.2*X[:,1] + 0.1*np.random.rand(num_data)

'''x=6*np.random.rand(m,1)-2
y1=x**3+-3*x**2-3*x+1
y=x**3+-3*x**2-3*x+7*np.random.randn(m,1)
plt.scatter(x,y1,color='r')'''
print(np.shape(np.random.rand(num_data,1)))
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0],X[:,1] , y, c='r', marker='o', s=15, cmap='Greens')
#plt.scatter(X,y)
#x=np.reshape(X,(-1,2))
```

```python
import matplotlib.pyplot as plt
import tensorflow as tf
model = tf.keras.models.Sequential([
  tf.keras.Input(shape=(2)),
  tf.keras.layers.Dense(64, activation='relu'),
  tf.keras.layers.Dense(32, activation='relu'),
  #tf.keras.layers.Dense(16, activation='relu'),
  tf.keras.layers.Dense(1, activation='linear')
])
model.compile(optimizer='adam',
              loss='mse')
model.fit(X, y, epochs=300)
```

```python
model.summary()
```

```python
def plot_predictions(X, y, g,model):
    fig = plt.figure(figsize=(6, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(X[:,0],X[:,1] , g, c='g', marker='o', s=15)
    x_test1=np.reshape(np.array([2.1,-3.2]),(1,-1))
    y_test1=model.predict(x_test1)
    ax.scatter(x_test1[0,0],x_test1[0,1] , y_test1, c='y', marker='o', s=100)
    x_test2=np.reshape(np.array([-3.2,2.1]),(1,-1))
    y_test2=model.predict(x_test2)
    ax.scatter(x_test2[0,0],x_test2[0,1] , y_test2, c='b', marker='o', s=100)

    
    plt.show()
```

```python
g=model.predict(X)
plot_predictions(X, y, g,model)
```
