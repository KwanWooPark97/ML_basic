```python
from sklearn.datasets import load_diabetes
import pandas as pd
import numpy as np
diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names, dtype=np.float32)
df.tail(3)
```

```python
X1 = df.iloc[:-100, [0,2,3,4,5]].values
X1_test=df.iloc[-100:, [0,2,3,4,5]].values
X2 = df.iloc[:-100,[2,3,6,7,8]].values
X2_test=df.iloc[-100:, [0,2,3,4,5]].values
y = diabetes.target.astype(np.float32)
y_train=y[:-100]
y_test=y[-100:]
```

```python
import tensorflow as tf

input = tf.keras.layers.Input(shape=[5])
x1 = tf.keras.layers.Dense(256, activation='sigmoid')(input)
x2 = tf.keras.layers.Dense(128, activation='sigmoid')(x1)
x3 = tf.keras.layers.Dense(64, activation='sigmoid')(x2)
predictions = tf.keras.layers.Dense(1, activation='linear')(x3)

model = tf.keras.Model(inputs=input, outputs=predictions)

model.compile(optimizer='adam',
              loss='mse')
model.fit(X1, y_train, epochs=1000)
```

```python
input = tf.keras.layers.Input(shape=[5])
x1 = tf.keras.layers.Dense(256, activation='sigmoid')(input)
x2 = tf.keras.layers.Dense(128, activation='sigmoid')(x1)
x3 = tf.keras.layers.Dense(64, activation='sigmoid')(x2)
predictions = tf.keras.layers.Dense(1, activation='linear')(x3)

model1 = tf.keras.Model(inputs=input, outputs=predictions)

model1.compile(optimizer='adam',
              loss='mse')
model1.fit(X2, y_train, epochs=1000)
```

```python
model.evaluate(X1_test,y_test)
```

```python
model1.evaluate(X2_test,y_test)
```
