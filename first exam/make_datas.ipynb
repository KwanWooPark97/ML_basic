```python
import numpy as np
import numpy.random as rnd
import matplotlib.pyplot as plt
np.random.seed(42)

m=500

X=6*np.random.rand(m,1)-2
y=X**3+X-3*X**2-4*X+7*np.random.randn(m,1)

plt.scatter(X,y)
```

```python
import pandas as pd

arr=np.concatenate((X, y), axis=1)
df = pd.DataFrame(arr)
 
df.to_csv('mydata.csv', index=False,header = False)
```
